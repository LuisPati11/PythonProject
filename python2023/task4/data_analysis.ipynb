```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file = '../salaries_clean.csv'

data = pd.read_csv(file)
```

```python
data['annual_base_pay'] = data['annual_base_pay'].astype(float)
column_mean = data['annual_base_pay'].mean()  
print("Mean of salary:", round(column_mean, 2),'€')
```

```python
column_moda = data['location_name'].mode().values[0]  # Reemplaza 'columna' con el nombre de la columna deseada
print("Most repeated location:", column_moda)
```

```python
top_20_categories = data['location_name'].value_counts().nlargest(10)

top_20_categories.plot(kind='bar')
plt.xlabel('Locations')
plt.ylabel('Frequency')
plt.title('Bar Chart of locations (Top 10)')
plt.show()
```

```python
data.boxplot(column=['annual_base_pay'])
plt.ylabel('Values')
plt.title('Box Plot of column_name')
plt.show()
```
