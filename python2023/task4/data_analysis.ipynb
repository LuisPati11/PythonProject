```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file = '../salaries_clean.csv'

data = pd.read_csv(file)
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe(include='all')
```

```python
data = data.sort_values('total_experience_years')  # Sort data by experience

plt.figure(figsize=(20, 12))  # Adjust the size of the figure

plt.scatter(data['total_experience_years'], data['annual_base_pay'])
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Relationship between Experience and Salary')


# Delete the y-axis labels
plt.yticks('')


plt.show()
```

```python
top_20_categories = data['location_name'].value_counts().nlargest(10)

top_20_categories.plot(kind='bar')
plt.xlabel('Locations')
plt.ylabel('Frequency')
plt.title('Bar Chart of locations (Top 10)')
plt.show()
```

```python
# Count occurrences of each job_title
job_counts = data['job_title'].value_counts()

# Get the 10 most common jobs
top_10_jobs = job_counts.head(10)

# Filter data to include only the 10 most common jobs
filtered_data = data[data['job_title'].isin(top_10_jobs.index)]

# Create the histogram
plt.hist(filtered_data['job_title'], bins=10)
plt.xlabel('Jobs')
plt.xticks(rotation=90)
plt.ylabel('Vacancies')
plt.title('Histogram')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Conclusions about the dataset

### First conclusion:

For the scatter plot we can say that the conclusion that can be seen visually is that the more years of experience you have in the sector, the higher the salary you earn. It is perfectly reflected in our graph

### Second conclusion:

In our bar chart showing the locations where we can find the most vacancies, we can draw the conclusion that the USA is the country with the most vacancies. From the top 10 locations with the highest number of vacancies, 9 of them belong to the USA, all except London.

### Third conclusion:

In our last graph, the histogram. We can conclude that the enginer software job is well above any other job in our dataset as it is more than twice as high as the second place job.
